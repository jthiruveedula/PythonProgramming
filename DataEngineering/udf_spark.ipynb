```
from pyspark.sql import SparkSession
dbutils.widgets.text("file_location", "/uploads/data", "Upload Location")
dbutils.widgets.dropdown("formatType", "csv", [x for x in ['csv','parquet','text','avro','orc']])
dbutils.widgets.dropdown("optionTrue", "True", [x for x in ['True','False']])
#[csv,parquet,text,avro,orc]
```

```
spark = SparkSession.builder.appName("filetester").getOrCreate()
```

```
df1=spark.read.format(dbutils.widgets.get("formatType")).option('header',dbutils.widgets.get("optionTrue")).option('inferSchema',dbutils.widgets.get("optionTrue")).load(dbutils.widgets.get("file_location"))
```

```
import re
def charRemoval(s):
  s=re.sub('[^0-9]+', '', s)
  return s
spark.udf.register("charRemoval", charRemoval)
```

```
import re
from  pyspark.sql.types import IntegerType
def charRemoval(s):
  s=re.sub('[^0-9]+', '', s)
  return int(s)
spark.udf.register("charRemoval", charRemoval,IntegerType())
```
